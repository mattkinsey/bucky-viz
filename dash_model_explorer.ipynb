## Bucky Covid19 Forecast Model Explorer 

Bucky-Forecast - https://github.com/mattkinsey/bucky-forecasts/<br>
ReichLab Covid-Forecast - https://github.com/reichlab/covid19-forecast-hub<br>

This code can import CSV model outputs and plot the combined historical and predicted Time Series.  There are feature selection dropdown toggles and confidence interval parameter adjustments.  The check boxes allow comparison between model runs.

1.  The latest Data can be loaded by cloning the ReichLab covid19-forecast repo and the Bucky-Forecast repo.
2.  The paths to the folders where these data files are pulled down will need to be established.
3.  If the schema of these files changes, this code will need to be repointed.
4.  When running the Dash code cell, be aware that this will start a server process that will need to be interrupted in order to end.

This version includes a compiled demo data object that can be used to skip the data download step.  Circa August 2022.
